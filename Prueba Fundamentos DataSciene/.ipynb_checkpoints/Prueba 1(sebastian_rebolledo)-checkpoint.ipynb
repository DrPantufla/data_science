```python
df_origin = pd.read_csv('income-db.csv')
```

### A continuacion realizare la recodificacion de las variables.

```python
df.workclass.value_counts()
```

```python
for columna in df.columns:
    df[columna] = np.where(df[columna]=='?', np.nan, df[columna])
```

```python
df.workclass.value_counts()
```

```python
df = df.replace(['Prof-specialty', 'Exec-managerial', 'Adm-clerical', 'Sales', 'Tech-support'], 'white-collar')
df = df.replace(['Craft-repair', 'Machine-op-inspct', 'Transport-moving', 'Handlers-cleaners', 'Farming-fishing',
                 'Protective-serv', 'Priv-house-serv'], 'blue-collar')
df = df.replace(['Other-service', 'Armed-Forces'], 'others')
```

```python
df = df.replace(['Federal-gov'], 'federal-gov')
df = df.replace(['State-gov','Local-gov'], 'state-level-gov')
df = df.replace(['Self-emp-inc', 'Self-emp-not-inc'], 'self-employed')
df = df.replace(['Never-worked', 'Without-pay'], 'unemployed')
```

```python
df = df.replace(['Preschool'],'preschool')
df = df.replace(['1st-4th', '5th-6th'], 'elementary-school')
df = df.replace(['7th-8th', '9th', '10th','11th', '12th', 'HS-grad'], 'high-school')
df = df.replace(['Assoc-voc', 'Assoc-acdm', 'Some-college'], 'college')
df = df.replace(['Bachelors', 'Masters', 'Prof-school', 'Doctorate'], 'university')
```

```python
df = df.replace(['Married-civ-spouse', 'Married-spouse-absent', 'Married-AF-spouse'], 'married')
df = df.replace(['Divorced'], 'divorced')
df = df.replace(['Separated'], 'separated')
df = df.replace(['Widowed'], 'widowed')
```

```python
df['marital-status'].value_counts()
```

```python
df.income = np.where(df['income']=='>50K', 1, 0)
```

```python
df = df.replace(['United-States', 'Puerto-Rico', 'Canada', 'Cuba', 'Honduras', 'Mexico',
                              'Dominican-Republic', 'Jamaica', 'Ecuador', 'Haiti', 'Columbia', 'Guatemala', 
                              'Nicaragua', 'El-Salvador', 'Trinadad&Tobago', 'Peru'], 'america') 
df = df.replace(['England','Germany', 'Greece', 'Poland', 'France', 'Hungary', 'Holand-Netherlands', 'Yugoslavia',
           'Scotland', 'Ireland', 'Portugal', 'Italy'], 'europe')
df = df.replace(['Outlying-US(Guam-USVI-etc)'], 'oceania')
df = df.replace(['Cambodia', 'India', 'Japan', 'South', 'China', 'Iran', 'Philippines', 'Vietnam', 'Laos',
            'Taiwan', 'Thailand', 'Hong'], 'asia')
```

```python
df
```

```python
df['native-country'].value_counts()
```

```python
df = df.rename(columns = {'native-country':'region', 'occupation':'collars', 'education':'educ_recod',
                          'workclass':'workclass_recod','marital-status':'civstatus'})
```

```python
df
```

### Tabla solo con numeros

```python
n_df=df.copy() 
```

```python
work=preprocessing.LabelEncoder().fit(n_df.workclass_recod.value_counts().index)
n_df.workclass_recod=n_df.workclass_recod.replace(n_df.workclass_recod.value_counts().index,
                                                  work.transform(n_df.workclass_recod.value_counts().index))
```

```python
educ=preprocessing.LabelEncoder().fit(n_df.educ_recod.value_counts().index)
n_df.educ_recod=n_df.educ_recod.replace(n_df.educ_recod.value_counts().index,
                                                  educ.transform(n_df.educ_recod.value_counts().index))
```

```python
civ=preprocessing.LabelEncoder().fit(n_df.civstatus.value_counts().index)
n_df.civstatus=n_df.civstatus.replace(n_df.civstatus.value_counts().index,
                                                  civ.transform(n_df.civstatus.value_counts().index))
```

```python
coll=preprocessing.LabelEncoder().fit(n_df.collars.value_counts().index)
n_df.collars=n_df.collars.replace(n_df.collars.value_counts().index,
                                                  coll.transform(n_df.collars.value_counts().index))
```

```python
relation_ship=preprocessing.LabelEncoder().fit(n_df.relationship.value_counts().index)
n_df.relationship=n_df.relationship.replace(n_df.relationship.value_counts().index, 
                                            relation_ship.transform(n_df.relationship.value_counts().index))
```

```python
race_=preprocessing.LabelEncoder().fit(n_df.race.value_counts().index)
n_df.race=n_df.race.replace(n_df.race.value_counts().index, 
                            race_.transform(n_df.race.value_counts().index))
```

```python
n_df.gender=np.where(n_df.gender=='Female', 1, 0)
```

```python
region_=preprocessing.LabelEncoder().fit(n_df.region.value_counts().index)
n_df.region=n_df.region.replace(n_df.region.value_counts().index, 
                            region_.transform(n_df.region.value_counts().index))
```

```python
n_df=n_df.rename(columns={'capital-gain':'capital_gain','capital-loss':'capital_loss', 
                     'hours-per-week':'hours_per_week', 'educational-num':'educational_num'})
```

# Hito 4

#### Procedo a realizar el modelo de regresion logistica en Machine Learning utilizando las variables seleccionadas en hito anterior

```python
y_vec = n_df.loc[:, 'income']
x_mat = n_df.loc[:,['age', 'workclass_recod', 'fnlwgt', 'educational_num',
       'civstatus', 'collars', 'relationship', 'race', 'gender',
       'capital_gain', 'capital_loss', 'hours_per_week']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_mat, y_vec, test_size=.33, random_state=111)
```

```python
x_train_std = StandardScaler().fit_transform(x_train)
x_test_std =StandardScaler().fit_transform(x_test)
```

```python
regresion_logistica = LogisticRegression().fit(x_train_std, y_train)
```

```python
yhat=regresion_logistica.predict(x_test_std)
```

#### Matriz de confusion, utilizada para evaluar el desempeño del modelo realizando predicciones.

```python
matriz_confusion = confusion_matrix(y_test, yhat)
matriz_confusion
```

#### La misma Matriz de confusion, pero como tabla, lo que ayuda a su comprension.

```python
matriz_confusion_pd = pd.crosstab(y_test, yhat)
```

```python
matriz_confusion_pd
```

#### En base a lo observado en la matriz de confusion, podemos concluir que nuestro modelo, predice muy bien los 0, pero no asi con los valores iguales a 1.

#### A continuacion utilizare classification_report para obtener el valor de las metricas de Precision, Recall  y F1-Score

```python
print(classification_report(y_test, yhat))
```

- Al observar la tabla obtenidad se reafirma lo dicho al observar la matriz de confusión, el modelo predice muy bien los 0 pero no asi los 1.
- Si comparamos el valor de precision con el valor de realizar una prediccion azarosa(1/(cantidad de variable): 1/12 = 0.0833 podemos decir que nuestro modelo tiene un desempeño de un 77% superior prediciendo el valor 0 que una prediccion azaroza.
- Y al predecir el valor de 1, el modelo tiene un desempeño de un 62% superior que una prediccion azaroza.
- Ahora al observar el valor de Recall, nos dice la proporcion de positivos reales se logro identificar, observandose que para el valor 0 es de un 93% y para 0 es de un 52%, lo que se resumen, como lo mencione anteriormente, que el modelo predice bien el valor 0 pero no asi el valor 1.

```python
fa.curva_roc(y_test, yhat)
```

El grafico nos indica que nuestro modelo tiene capacidad predictiva al encontrarse la curva ROC sobre la bisectriz.

```python
roc_auc_score(y_test, yhat).round(2)
```

Utilizando roc_auc_score obtenemos el area bajo la curva ROC, con un valor de 0.73 nos indica que nuestro modelo tiene un desempeño de un 23% superior a un mecanismo de clasificacion aleatoria.

## Modelo con crossvalidation

#### Crossvalidation entregando el valor de F1

```python
m1_logit_cv_f1 = cross_val_score(LogisticRegression(),x_mat,y_vec, cv=10, scoring='f1').mean()
print(m1_logit_cv_f1)
```

#### Crossvalidation entregando el valor de recall

```python
m1_logit_cv_recall = cross_val_score(LogisticRegression(),x_mat,y_vec, cv=10, scoring='recall').mean()
print(m1_logit_cv_recall)
```

#### Crossvalidation entregando el valor de accuracy

```python
m1_logit_cv_acc = cross_val_score(LogisticRegression(),x_mat,y_vec, cv=10, scoring='accuracy').mean()
print(m1_logit_cv_acc)
```

#### Los valores obtenidos de Recall y F1 son bastante menores que los obtenidos en el modelo anterior, el unico con un valor similar fue accuracy.

### Reporte Modelo elegido

En base a los resultados obtenidos utilizando la regresion logistica con y sin cross validation, obtuve mejores valores en las metricas de Recall y F1 en el modelo sin cross validation. Accuracy se mantuvo similar en ambas.
Por lo que el mejor modelo segun lo comentado anteriormente es el de Regresion Logistica sin Cross validation.
