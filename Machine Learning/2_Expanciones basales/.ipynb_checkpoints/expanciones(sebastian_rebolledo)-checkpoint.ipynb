### Desafio 3

#### Al ser el vector objetivo una variable continua, he decidido utilizar LinearGam

```python
vec_objtivo = df.iloc[:, 8]
matrix_train = df.iloc[:,:8]
```

```python
x_train_pre, x_test_pre, y_train, y_test = train_test_split(matrix_train, vec_objtivo, test_size=0.3, random_state=1235)
```

#### El estandarizador que utilizare será StandardScaler

```python
scaler = StandardScaler().fit(x_train_pre)
```

```python
x_train = pd.DataFrame(scaler.transform(x_train_pre), columns = x_train_pre.columns)
```

```python
x_test = pd.DataFrame(scaler.transform(x_test_pre), columns = x_test_pre.columns)
```

#### A continuación, se realizara el modelo sin utilizar gridsearch.

```python
gam_pre = LinearGAM(s(0)+s(1)+s(2)+s(3)+s(4)+s(5)+s(6)+s(7), fit_intercept =True).fit(x_train, y_train)
```

```python
gam_pre.summary()
```

```python
def report_metrics(model, X_test, y_test):
    y_hat=model.predict(X_test)
    print(f'Test R^2 : {r2_score(y_test,y_hat).round(3)}')
    print(f'Test RMSE: {np.sqrt(mean_squared_error(y_test, y_hat)).round(3)}')
    print(f'Test MAE: {median_absolute_error(y_test, y_hat).round(3)}')
    
```

```python
report_metrics(gam_pre, x_test, y_test)
```

#### A continuación, se utilizara gridsearch para obtener el valor de los lamda 

```python
lams = np.logspace(-3,3,3)
```

```python
lams = [lams]*len(x_train.columns)
```

```python
gam = LinearGAM(s(0)+s(1)+s(2)+s(3)+s(4)+s(5)+s(6)+s(7), fit_intercept =True)
```

```python
gam.gridsearch(x_train.values, y_train.values, lam = lams)
```

```python
gam.summary()
```

```python
report_metrics(gam, x_test, y_test)
```

#### Al comparar las metricas de los 2 modelos realizados:
- Se observa un r2 casi igual en ambos.
- El RMSE y MAE es menor en el en modelo utilizando gridsearch.
- El valor del hiperparametro lamda en el modelo sin gridsearch es igual a 0.6 para las 8 columnas.
- El valor del hiperparametro lamda en el modelo con gridsearch es igual a 1 para las 7 primeras columnas y la ultima es igual a 0.001.

## Desafio 4

```python
fig, axs = plt.subplots(4, 2, figsize = (20,18))
titles= x_train.columns
for i, ax in enumerate(axs.flatten()):
    xx = gam.generate_X_grid(term=i)
    ax.plot(xx[:,i], gam.partial_dependence(term = i, X=xx))
    ax.plot(xx[:,i], gam.partial_dependence(term = i, X=xx, width=.95)[1], c='r', ls='--')
    ax.scatter(x_train[titles[i]], [0]*len(x_train[titles[i]]), marker = '|', alpha =.5)
    ax.set_title(titles[i])
```

Observaciones: 

- En el grafico **Cement** nos muestra que entre más cemento se agregue en la mezcla genera mas fuerza compresiva.
- En el grafico **Blast Furnace Slag**, se ve en el grafico que aumenta la fuerza compresiva entre más material se le agrega a la mezcla hasta un punto maximo y luego empieza a decaer entre mas material se le agregue.
- En el grafico **Fly Ash** se observa que la fuerza compresiva al inicio es mas variable en función de la cantidad utilizada en la mezcla, luego llega a su punto maximo de fuerza compresiva y posteriormente disminuye lentamente a medida que aumenta la cantidad.
- En el grafico **Water**, se observa que se necesita poca cantidad en la mezcla para llegar al maximo valor de fuerza compresiva, y si se le continua agregando agua disminuye la fuerza compresiva.
- En el grafico **Superplasticizer**, se observa que tiende a disminuir la fuerza compresiva a medida que se le agrega mas material.
- En el grafico **Coarse Aggregate**, se observa que se debe agregar bastante material en la mezcla para lograr que aumente la fuerza compresiva. 
- En el grafico **Fine Aggregate**, se observa que se debe agregar bastante material en la mezcla para aumentar la fuerza compresiva.
- En el grafico **Age(day)**, se observa que al pasar los dias el material aumenta su fuerza compresiva.
