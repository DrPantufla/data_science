### Ejercicio 2 - Evaluación de modelos individuales

#### Regresion Logistica

```python
X=df.drop('shares', axis=1)
y=df['shares']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=131)
```

```python
X_train_std = StandardScaler().fit_transform(X_train)
X_test_std = StandardScaler().fit_transform(X_test)
```

```python
log_regres = LogisticRegression().fit(X_train_std, y_train)
```

```python
y_hat_log = log_regres.predict(X_test_std)
```

```python
print(classification_report(y_test, y_hat_log))
```

#### Modelo KMeans

```python
kmeans = KMeans(n_clusters=2).fit(X_train_std, y_train)
```

```python
y_hat_km = kmeans.predict(X_test_std)
```

```python
print(classification_report(y_test, y_hat_km))
```

El modelo generado con Kmeans no se puede ocupar con VotingClassifier debido a que no es un modelo de clasificación, por lo que se cambiara por AdaBoostClassifier.

#### Arbol de Clasificación max_depth = 1

```python
dec_tree_one_depth = DecisionTreeClassifier(random_state=452, max_depth=1).fit(X_train, y_train)
```

```python
y_hat_tree_one = dec_tree_one_depth.predict(X_test)
```

```python
print(classification_report(y_test, y_hat_tree_one))
```

#### Arbol de Clasificación max_depth = 4

```python
dec_tree_four_depth = DecisionTreeClassifier(random_state=452, max_depth=4).fit(X_train, y_train)
```

```python
y_hat_tree_four = dec_tree_four_depth.predict(X_test)
```

```python
print(classification_report(y_test, y_hat_tree_four))
```

#### AdaBoostClassifier

```python
params_ada={'learning_rate': [0.01, 0.1, 0.5], 'n_estimators': [50,100, 500, 1000, 2000]}
ada_grid_cv = GridSearchCV(AdaBoostClassifier(), params_ada, n_jobs = -1).fit(X_train, y_train)
```

```python
ada_grid_cv.best_params_
```

```python
y_hat_adaboost =ada_grid_cv.best_estimator_.predict(X_test)
```

```python
print(classification_report(y_test, y_hat_adaboost))
```

#### Este modelo presenta mejores metricas que los 4 anteriores.

### Ejercicio 3 - Entrenamiento de Comité

```python
estimators = [('Logistic Regression', LogisticRegression(random_state=1)),
              ('Adaboostclassifier', AdaBoostClassifier(learning_rate=0.1, n_estimators=2000, random_state=1)),
              ('DecisionTreeClassifier-max_depth: 1', DecisionTreeClassifier(random_state=1, max_depth=1)),
              ('DecisionTreeClassifier-max_depth: 4', DecisionTreeClassifier(random_state=1, max_depth=4))
             ]
```

#### Se cambio el metodo Kmeans por AdaBoostClassifier ya que el modulo VotingClassifier solo acepta modelos de clasificación

```python
estimators[1]
```

```python
voting_classifier = VotingClassifier(estimators=estimators).fit(X_train_std, y_train)
```

```python
y_hat_vc = voting_classifier.predict(X_test_std)
```

```python
print(classification_report(y_test, y_hat_vc))
```

#### No se observa una mejora con respecto a los modelos anteriores.

### Ejercicio 4 - Calibración de Comité con Ponderadores

Esquema de ponderación 1: Por porcentajes.
Los mejores modelos tendrán mayores valores de peso. El orden a continuacion seria:
- Regresion logistica, pondera = 0.255.
- Arbol de clasificacion con un max_depth = 1, pondera = 0.231.
- Arbol de clasificacion con un max_depth = 4, pondera = 0.245.
- AdaBoostClassifier, pondera = 0.27.
- weigths = [0.255,0.231, 0.247, 0.27]
Se observa que no hay una gran diferencia porcentual entre los modelos.

Esquema de ponderación 2:
Los mejores modelos tendrán mayores valores de peso. El orden a continuacion seria:
- Regresion logistica, pondera = 1.5.
- Arbol de clasificacion con un max_depth = 1, pondera = 0.5.
- Arbol de clasificacion con un max_depth = 4, pondera = 1.
- AdaBoostClassifier, pondera = 2.
- weigths = [1.5, 0.5, 1, 2]

```python
sum = 0
for i in [0.254,0.231, 0.245, 0.27]:
    sum+=i
print(sum)
```

### Modelo 1 con ponderaciones. 

```python
vc_weigths_1 = VotingClassifier(estimators=estimators, weights=[0.255,0.231, 0.247, 0.27]).fit(X_train_std, 
                                                                                               y_train)
```

```python
y_hat_vc_weigths_1 = vc_weigths_1.predict(X_test_std)
```

```python
print(classification_report(y_test, y_hat_vc_weigths_1))
```

Se obtiene un modelo con menor desempeño que el obtenido con AdaBoostClassifier y que el modelo sin ponderaciones, por lo que se deben cambiar los ponderadores.

### Modelo 2 con ponderaciones. 

```python
vc_weigths_2 = VotingClassifier(estimators=estimators, weights=[3, 0.5, 1, 1.5]).fit(X_train_std, y_train)
```

```python
y_hat_vc_weigths_2 = vc_weigths_2.predict(X_test_std)
```

```python
print(classification_report(y_test, y_hat_vc_weigths_2))
```

#### Se logra una pequeña mejora comparado con el modelo de VotingClassifier sin ponderadores, pero sigue siendo peor que las metricas obtenidas con el modelo de AdaBoostClassifier.
