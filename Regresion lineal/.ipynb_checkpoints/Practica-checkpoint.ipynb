```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
attr_mat = df.drop(['earn'],axis = 1)
```

```python
x_mat_train = attr_mat[:-30]
x_mat_test = attr_mat[-30:]

y_train = df['earn'][:-30]
y_test = df['earn'][-30:]
```

```python
x_mat_train.shape
```

```python
x_mat_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
modelo_sklearn = linear_model.LinearRegression().fit(x_mat_train, y_train)
```

```python
modelo_sklearn.coef_
```

```python
attr_mat.columns
```

```python
modelo_sklearn.intercept_
```

```python
earn_yhat = modelo_sklearn.predict(x_mat_test)
```

```python
earn_yhat
```

```python
mse_modelo = mean_squared_error(y_test, earn_yhat).round(1)
r2_modelo = r2_score(y_test, earn_yhat).round(1)

print(f'MSE: {mse_modelo}\nR2: {r2_modelo}')
```
