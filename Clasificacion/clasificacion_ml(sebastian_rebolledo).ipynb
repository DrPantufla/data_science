### Desafío 3: Modelación econométrica

```python
df.student = np.where(df.student == 'Yes', 1, 0)
```

```python
df.default = np.where(df.default == 'Yes', 1, 0)
```

```python
df
```

```python
model_log = smf.logit('default ~ balance+income+student', df).fit()
```

```python
model_log.summary2()
```

```python
model_log2 = smf.logit('default ~ balance+student', df).fit()
```

```python
model_log2.summary2()
```

Los atributos que conforman el modelo tienen una significancia del 99%, lo que permite rechazar la hipotesis nula.

### Desafío 4: Entrenar 

```python
X_train, X_test, y_train, y_test =train_test_split(df.loc[:,['student', 'balance']], df['default'], 
                                                   test_size=.33, random_state= 5)
```

```python
X_train_std =StandardScaler().fit_transform(X_train)
X_test_std =StandardScaler().fit_transform(X_test)
```

```python
modelo_moroso = LogisticRegression().fit(X_train_std, y_train)
```

```python
modelo_moroso.coef_
```

```python
yhat = modelo_moroso.predict(X_test_std)
```

```python
yhat[:20]
```

```python
modelo_moroso_confusion = confusion_matrix(y_test, yhat)
```

```python
modelo_moroso_confusion
```

```python
modelo_moroso_report = classification_report(y_test, yhat)
```

```python
print(modelo_moroso_report)
```

```python
yhat=modelo_moroso.predict_proba(X_test)[:,1]
```

```python
false_positive, true_positive, threshold = roc_curve(y_test, yhat)
```

```python
plt.title('Curva Roc')
plt.plot(false_positive,  true_positive, lw =1)
plt.plot([0,1],ls='--', lw=1)
plt.plot([0,0], [1,0], c='green', lw=3), plt.plot([1,1], c='green', lw=3)
plt.ylabel('Verdaderos Positivos')
plt.xlabel('Falsos Positivos')
plt.show()
```

- La Curva ROC nos muestra que los verdaderos positivos son mayores que los falsos negativos, lo que indica que nuestro modelo tiene capacidad predictiva adecuada.
- Accuracy nos indica que el modelo identifica el 98% de los 0, y un 85% de los 1.
- Recall indica que el modelo identifico correctamente el 100% de los positivos reales de 0, y 29% de los reales positivos de 1.
