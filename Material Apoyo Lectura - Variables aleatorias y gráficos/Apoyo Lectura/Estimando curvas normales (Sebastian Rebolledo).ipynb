**Desafio 5**

```python
from scipy.stats import norm
```

```python
def densidad_probabilidad(df, variable):
    puntaje_z = []
    mean_variable = df[variable].dropna().mean()
    std_variable = df[variable].dropna().std()
    for valor in df[variable].dropna():
        new_valor = (valor-mean_variable)/std_variable
        puntaje_z.append(new_valor)
    
    puntaje_z.sort()
    plt.plot(puntaje_z, norm.pdf(puntaje_z, 0,1))
    return 
```

```python
densidad_probabilidad(df, 'imf_pop')
```

```python
densidad_probabilidad(df, 'gle_cgdpc')
```

```python
densidad_probabilidad(df, 'wdi_exph')
```
