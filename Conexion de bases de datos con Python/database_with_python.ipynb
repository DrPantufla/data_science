```python
def str_to_int_bool(list_str):
    list_float = [float(i) for i in list_str]
    list_int = [int(i) for i in list_float]
    list_bool = []
    for i in list_int:
        if i == 0:
            list_bool.append('0')
        elif i == 1:
            list_bool.append('1')
        else:
            list_bool.append(i)
       
    return list_bool
```

```python
with open('./train_cupid.csv', 'r') as train:
    reader = csv.reader(train, delimiter=',', quotechar=',', quoting=csv.QUOTE_MINIMAL)
    next(reader)
    for row in reader:
        row=str_to_int_bool(row)
        cur.execute(insert_train, row)
```

```python
X_test = df_test.drop(['single','seeing_someone','available'], axis=1)
y_test_1 = df_test['single']
y_test_2 = df_test['seeing_someone']
y_test_3 = df_test['available']
```

```python
def prediction_model(model, x_train, y, x_test, y_test,print_report=False ):
    X_train_std = StandardScaler().fit_transform(X_train)
    model_train = model().fit(X_train_std, y)

    y_hat = model_train.predict(X_test)
    report = classification_report(y_test, y_hat)
    y_hat_test = pd.Series(y_hat).rename('Class Prediction')
    df_concat = pd.concat([X_test, y_hat_test], axis=1)
    if print_report:
        print(report)
    else:
        return df_concat
```

```python
engine = create_engine('postgresql://postgres:1234@localhost/rebolledo_sebastian')
```

```python
def table_in_postgres(query, modelo, model_name, engine, vector, query_name):
    df_comb = modelo.groupby(query)['Class Prediction'].mean()
    tmp_table = pd.DataFrame(df_comb)
    return tmp_table.to_sql(query_name+'_'+vector[:-4]+'_'+model_name.__name__, con = engine)
```

```python
for models in [GradientBoostingClassifier, AdaBoostClassifier, RandomForestClassifier, DecisionTreeClassifier, BernoulliNB, 
               LogisticRegression, SVC]:
    for vec_obj in ['single','seeing_someone','available']:
        query = ['atheism','asian', 'employed', 'pro_dogs', 'chinese']

        modelo = prediction_model(models, X_train, df_train[vec_obj], X_test, df_test[vec_obj])
        
        table_in_postgres(query, modelo, models, engine, vec_obj, 'query1')
        
```
